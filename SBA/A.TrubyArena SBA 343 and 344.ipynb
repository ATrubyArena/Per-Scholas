```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

def extract_from_csv(): 
    dataframe = pd.read_csv("woocommerce-product-export.csv") 
    return dataframe 
df = extract_from_csv()
df.info()
df.describe()
print(df.head())
print(df.tail())

for month, profit in zip(df["month_number"], df["total_profit"]):
    print(f"Month: {month}, Total Profit: {profit}")

plt.figure(figsize=(16, 10))
plt.bar(df["month_number"], df["total_profit"], color='green')
plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")

plt.figure(figsize=(16, 10))
    
    
plt.plot(
    df["month_number"],
    df["total_profit"],
    linestyle=':',
    color='red',            
    marker='o',      
    markerfacecolor='black',
    label='Total Profit',
    linewidth=3 
)

plt.xlabel("Month number")
plt.ylabel("Total profit")
plt.title("Company profit per month")
plt.legend(loc='lower right')
plt.xticks(df["month_number"])
plt.grid(True)
plt.tight_layout()
plt.show()

product_columns = [col for col in df.columns if col in ['bathingsoap']]

plt.figure(figsize=(16, 10))

for product in product_columns:
    plt.scatter(
        df['month_number'],
        df[product],
        marker='o',
        label=product,
        linewidth=2
)

plt.xlabel("Month number")
plt.ylabel("Units sold")
plt.title("Monthly sales data for each product")
plt.legend()
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```
